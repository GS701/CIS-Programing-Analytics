```python
# requests is imported as it a vital component for the web scraping process, allowing the user to send http requests
import requests
# json is imported to as it a vital component for effectively handling and analyzing the data received from APIs
import json
# pandas is imported as it a vital component for dataframe creation and manipulation as well as data analysis through dataframes
import pandas as pd
```

```python
# The API key is assigned to the variable named key
key = "OQ3KE194BQ0W29RV"

# This is a list with a total of ten stocks assigned to the variable named stock_list so that ten request can be made to the API
stock_list = ['WMT', 'AMZN', 'NVDA', 'AAPL', 'TSLA', 'GOOG', 'META', 'AMD', 'MSFT', 'SONY']
```

```python
# The api_data function takes one parameter a list of stocks and returns a dataframe of the collected infromation from the API
def api_data(list):
  stocks = [] # Creates a empty list named stocks
  date = [] # Creates a empty list named date
  open_value = [] # Creates a empty list named open_value
  high_value = [] # Creates a empty list named high_value
  low_value = [] # Creates a empty list named low_value
  close_value = [] # Creates a empty list named close_value
  volume_value = [] # Creates a empty list named volume_value

  for stock_symbol in list: # Establishes a for statement that and goes through each element of the stock list
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={stock_symbol}&apikey={key}' # The url link of the API as f-string to utilize the variable named key as well the element from for statement and this is assigned to the variable named url
    data = requests.get(url) # An http get request is sent to the url variable and assigns the response to the variable named data
    raw_json = data.json() # The response from data variable and converts it from  Json fromat into dictionary and assigns it to the variable raw_json

    Symbol = raw_json['Meta Data']['2. Symbol'] # This extracts the value associated with the key '2. Symbol' from a nested dictionary within the raw_json dictionary and assigns this to a variable named Symbol
    stocks.append(Symbol) # This appends the Symbol variable to the empty list stocks

    Recent_date = raw_json['Meta Data']['3. Last Refreshed'] # This extracts the value associated with the key '3. Last Refreshed' from a nested dictionary within the raw_json dictionary and assigns this to a variable named Recent_date
    date.append(Recent_date) # This appends the Recent_date variable to the empty list date

    open = raw_json['Time Series (Daily)'][Recent_date]['1. open'] # This extracts the value associated with the key '1. open' from the nested dictionary within the raw_json dictionary, using the Recent_date variable as a key to access that dictionary, and assigns it to the variable named open.
    open_value.append(float(open)) # This appends the open variable as float to the empty list open_value

    high = raw_json['Time Series (Daily)'][Recent_date]['2. high'] # This extracts the value associated with the key '2. high' from the nested dictionary within the raw_json dictionary, using the Recent_date variable as a key to access that dictionary, and assigns it to the variable named high.
    high_value.append(float(high)) # This appends the high variable as float to the empty list high_value

    low = raw_json['Time Series (Daily)'][Recent_date]['3. low'] # This extracts the value associated with the key '3. low' from the nested dictionary within the raw_json dictionary, using the Recent_date variable as a key to access that dictionary, and assigns it to the variable named low.
    low_value.append(float(low)) # This appends the low variable as float to the empty list low_value

    close = raw_json['Time Series (Daily)'][Recent_date]['4. close'] # This extracts the value associated with the key '4. close' from the nested dictionary within the raw_json dictionary, using the Recent_date variable as a key to access that dictionary, and assigns it to the variable named close.
    close_value.append(float(close)) # This appends the close variable as float to the empty list close_value

    volume = raw_json['Time Series (Daily)'][Recent_date]['5. volume'] # This extracts the value associated with the key '5. volume' from the nested dictionary within the raw_json dictionary, using the Recent_date variable as a key to access that dictionary, and assigns it to the variable named volume.
    volume_value.append(float(volume)) # This appends the volume variable as float to the empty list volume_value

# This establishes a dataframe named data with key representing the infromation gathered with values that are the lists that were created earlier
  data = {
      'Stock Ticker': stocks,
      'Recent Date': date,
      'Open': open_value,
      'High': high_value,
      'Low': low_value,
      'Close': close_value,
      'Volume': volume_value
  }

  df = pd.DataFrame(data) # Dataframe is created assigning it the variable named df and uses the data dictionary to create it
  df.to_csv('stock_data.csv', index=False) # The dataframe df is exported into a csv file without the index and is named stock_data
  return df
```

```python
# This calls the function api_data and takes the stock_list as the parameter assigns it to the variable named df_api
df_api = api_data(stock_list)
# Displays the dataframe df_api
df_api
```
